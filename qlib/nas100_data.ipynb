```python
import qlib
# region in [REG_CN, REG_US]
from qlib.constant import REG_CN, REG_US
from qlib.data import D

provider_uri = "~/.qlib/qlib_data/us_data"  # target_dir
qlib.init(provider_uri=provider_uri, region=REG_US)
```

```python
# 加载交易日历
calendar = D.calendar(start_time='2010-01-01', end_time='2017-12-31', freq='day')
print(calendar[:2])  # 打印前两个日期
```

```python
# 获取纳斯达克 100 成分股
instruments = D.instruments(market='nasdaq100')
print(D.list_instruments(instruments=instruments, start_time='2010-01-01', end_time='2017-12-31', as_list=True)[:6])
```

```python
# 计算Alpha158特征（因子）和标签（收益）
from qlib.contrib.data.handler import Alpha158

# 设置日期、股票池等参数
data_handler_config = {
    "start_time": "2019-12-02", # 开始时间
    "end_time": "2024-12-31",   # 结束时间
    "fit_start_time": "2019-12-02", # 训练集开始时间
    "fit_end_time": "2020-06-30",   # 训练集结束时间
    "instruments": instruments,  # 请确保 instruments 已定义
}

# 初始化 Alpha158 数据处理器
h158 = Alpha158(**data_handler_config)

# 获取列名（因子名称）
print(h158.get_cols())
```

```python
print(len(h158.get_cols()))
```

```python
Alpha158_nasdaq100_feats = h158.fetch(col_set='feature')
Alpha158_nasdaq100_feats
```

```python
# 保存为 CSV 文件
csv_file_path = "alpha158_nasdaq100_feats.csv"  # 保存路径
Alpha158_nasdaq100_feats.to_csv(csv_file_path, index=True)  # 保存索引

print(f"数据已保存到: {csv_file_path}")
```
